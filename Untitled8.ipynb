```python
# Zomato EDA Project

#  Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#  Load Dataset
df = pd.read_csv("/content/Zomato-data-.csv")

#  Initial Exploration
print(df.head())
print(df.shape)
print(df.info())

#  Data Cleaning

# Drop unnecessary columns
# df = df.drop(['url', 'dish_liked', 'phone'], axis=1) # Removed this line as columns do not exist

# Remove duplicates
df.drop_duplicates(inplace=True)

# Remove null values
df = df.dropna(how='any')

# Clean "rate" column (remove "/5" and convert to float)
df['rate'] = df['rate'].apply(lambda x: str(x).split('/')[0].strip())
df = df[df['rate'] != 'NEW']
df = df[df['rate'] != '-']
df['rate'] = df['rate'].astype(float)

# Remove ',' from 'approx_cost(for two people)'
# df['approx_cost(for two people)'] = df['approx_cost(for two people)'].str.replace(',', '') # Removed this line as it's not needed for int type
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].astype(float)

# Rename for simplicity
df.rename(columns={
    'approx_cost(for two people)': 'cost',
    'listed_in(type)': 'type',
    'listed_in(city)': 'city'
}, inplace=True)

#  EDA & Visualizations

plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='online_order', palette='Set2')
plt.title("Online Ordering Availability")
plt.show()

plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='book_table', palette='Set3')
plt.title("Table Booking Availability")
plt.show()

# Top Locations
# plt.figure(figsize=(12, 6)) # Removed this block as 'location' column does not exist
# top_locations = df['location'].value_counts().head(10)
# sns.barplot(x=top_locations.index, y=top_locations.values, palette='viridis')
# plt.title("Top 10 Locations by Number of Restaurants")
# plt.xticks(rotation=45)
# plt.show()

# Online order vs Ratings
plt.figure(figsize=(8, 4))
sns.boxplot(x='online_order', y='rate', data=df, palette='Set1')
plt.title("Online Order vs Ratings")
plt.show()

# Cost vs Rating
plt.figure(figsize=(8, 4))
sns.scatterplot(x='cost', y='rate', data=df, alpha=0.5)
plt.title("Cost vs Rating")
plt.show()

# Cuisines Analysis (Top 10)
# plt.figure(figsize=(10, 4)) # Removed this block as 'cuisines' column does not exist
# top_cuisines = df['cuisines'].value_counts().head(10)
# sns.barplot(y=top_cuisines.index, x=top_cuisines.values, palette='coolwarm')
# plt.title("Top 10 Cuisines")
# plt.xlabel("Number of Restaurants")
# plt.ylabel("Cuisines")
# plt.show()

# Ratings by Type
plt.figure(figsize=(10, 4))
sns.boxplot(data=df, x='type', y='rate')
plt.xticks(rotation=45)
plt.title("Rating Distribution by Restaurant Type")
plt.show()

# Correlation Heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(df[['rate', 'cost', 'votes']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

#  Insights (as prints)
print("Top 5 Restaurant Chains:")
print(df['name'].value_counts().head(5))

# print("\nLocations with Highest Average Ratings:") # Removed this block as 'location' column does not exist
# print(df.groupby('location')['rate'].mean().sort_values(ascending=False).head(5))

# print("\nTop Rated Cuisines:") # Removed this block as 'cuisines' column does not exist
# print(df.groupby('cuisines')['rate'].mean().sort_values(ascending=False).head(5))
```
